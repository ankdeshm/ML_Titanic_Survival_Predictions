```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn import metrics


# Read data from scv files and drop index
Test = pd.read_csv('../input/titanic/test.csv', index_col=0)
Train = pd.read_csv('../input/titanic/train.csv', index_col=0)
gender = pd.read_csv('../input/titanic/gender_submission.csv', index_col=0)
```

```python
# Data Cleaning
# Drop unnecessary colums
to_drop = ['Name',
           'Ticket',
           'Fare',
           'Cabin']

Train.drop(to_drop, inplace=True, axis=1)
Test.drop(to_drop, inplace=True, axis=1)
```

```python
# Check for missing data
print(Train[:].isnull().sum())
print(Test[:].isnull().sum())
```

```python
# Replacing missing values of 'Age' with mean 'Age' in the dataset
mean_age_train = Train["Age"].mean()
Train["Age"].fillna(mean_age_train, inplace = True)

# Replacing missing values of 'Embarked' with where the most passengers embarked i.e by finding mode
mode_emb_train = Train["Embarked"].mode()[0]
Train["Embarked"].fillna(mode_emb_train, inplace = True)

# Replacing missing values of 'Age' with mean 'Age' in the dataset
mean_age_test = Test["Age"].mean()
Test["Age"].fillna(mean_age_test, inplace = True)

# Check for missing data
print(Train[:].isnull().sum())
print(Test[:].isnull().sum())
```

```python
# For regression we will need numeric data
# Hence in 'Sex', replacing 'male' = 1 and 'female' =0 
# In 'Embarked', replacing 'S' = 1, 'C' = 2 and 'Q' = 3

Train["Sex"].replace({"male": "1", "female": "0"}, inplace=True)
Train["Embarked"].replace({"S": "1", "C": "2", "Q": "3"}, inplace=True)
print(Train)

Test["Sex"].replace({"male": "1", "female": "0"}, inplace=True)
Test["Embarked"].replace({"S": "1", "C": "2", "Q": "3"}, inplace=True)
print(Test)
```

```python
X_train = Train.drop(columns=['Survived'])
y_train = Train['Survived']

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(Test)
y_pred = pd.Series(y_pred)
y_pred
```

```python
# accuracy_score: finding accuracy of s model
from sklearn.metrics import accuracy_score

y_actual = gender['Survived']
score = accuracy_score(y_actual, y_pred)
print(score)
```

```python
csv_input = pd.read_csv('../input/titanic/gender_submission.csv', index_col=0)
csv_input['Survived'] = y_pred
csv_input.to_csv('Titanic_Survival_Prediction.csv', index=False)
```
